### 선형 회귀의 정의

```python
# study_time = [2,4,6,8]
# score =      [81,93,91,97]
```

```python
import numpy as np
```

```python
X_y = np.array([
    [2,81],
    [4,93],
    [6,91],
    [8,97]
]) 
```

```python
X_y[:, 0]
```

```python
study_mean = X_y[:, 0].mean()
```

```python
score_mean = X_y[:, 1].mean()
```

```python
# x편차
X_y[:, 0] - study_mean
```

```python
# y편차
X_y[:, 1] - score_mean
```

```python
(X_y[:, 0] - study_mean) * (X_y[:, 1] - score_mean)
```

```python
sum((X_y[:, 0] - study_mean) * (X_y[:, 1] - score_mean))
```

```python
# 기울기
a = sum((X_y[:, 0] - study_mean) * (X_y[:, 1] - score_mean)) / \
    sum((X_y[:, 0] - study_mean)**2)
```

```python
# y절편 intercept
intercept = score_mean - (study_mean * a)
```

```python
# 평균 제곱 오차
import numpy as np
```

```python
fake_a = 3
fake_b = 76
```

```python
def predict(x):
    return fake_a * x + fake_b
```

```python
predict_result = []
```

```python
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])

for i in range(len(x)):
    predict_result.append(predict(x[i]))
    print(f"공부시간 {x[i]:.2f} " , end="") 
    print(f"실제점수 {y[i]:.2f} " , end="") 
    print(f"예측시간 {predict(x[i]):.2f}") 
    
n = len(x)

def mse(y, y_pred):
    return (1/n)* sum((y-y_pred)**2)

print("평균제곱 오차:" +  str(mse(y, predict_result)))
```
