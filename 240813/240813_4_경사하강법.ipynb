```python
import numpy as np
```

```python
x = np.array([2,4,6,8])
y = np.array([81,93,91,97])
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y, color='r')
plt.plot()
```

```python

plt.scatter(x, y, color='b', marker='o')
plt.plot()
```

```python
#   (parameter, weight)
```

```python
a, b = 0 , 0
```

```python
learning_rate = 0.03 # 
```

```python
epochs = 2001
```

```python
n = len(x)
```

```python
for i in range(epochs) :
    y_pred = a*x+ b
    error = y-y_pred
    
    a_diff = (2/n) * sum(-x* (error))
    b_diff = (2/n) * sum(-(error))
    
    a = a - learning_rate * a_diff
    b = b - learning_rate * b_diff
    
    if i % 100  == 0:
        print(f"result ::: epoch={i:.0f} ,  :{a:.2f},  : {b:.2f}")
```
