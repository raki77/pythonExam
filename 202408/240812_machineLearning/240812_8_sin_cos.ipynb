```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, color='orange', linewidth=2)
plt.xlabel('X')
plt.ylabel('Sin')
plt.title('LINE PLOT')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)

fig0 = plt.figure()
axes0 = fig0.add_axes([0,0,1,1])
axes0.plot(x, y, color='red', linestyle='-')
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')

plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np

fig0 = plt.figure()
axes0 = fig0.add_axes([0,0,1,1])
axes0.plot(x, y, color='purple', linestyle='-')
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')

axes0.set_xlim([0,5])
axes0.set_ylim([-2,2])

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig0 = plt.figure( figsize=(10, 1))
axes0 = fig0.add_axes([0,0,1,1])
axes0.plot(x, y, color='red', linestyle='-')

axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig0 = plt.figure( figsize=(5,2), dpi=100)
axes0 = fig0.add_axes([0,0,1,1])

axes0.plot(x, y, color='orange', linestyle='--', label='Sin')
axes0.plot(x, z, color='purple', linestyle='-.', label='Cos')

axes0.legend(loc=0)
axes0.set_xlabel('X')
axes0.set_ylabel('Y')
axes0.set_title('LINE PLOT')

plt.show()
```
