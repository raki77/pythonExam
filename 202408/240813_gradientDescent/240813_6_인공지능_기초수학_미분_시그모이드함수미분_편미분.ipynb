```python
import sympy as sp
```

```python
x, y = sp.symbols('x y')
a =  sp.symbols('a')
```

$ f(x, y) = x^2 + yx + a $

```python
f = x**2 + y*x + a
```

$\frac {∂f}{∂f}$   편미분이 'partial'이다.

```python
# x에 대한 편미분
sp.diff(f,x)
```

```python
# y에 대한 편미분
sp.diff(f,y)
```
