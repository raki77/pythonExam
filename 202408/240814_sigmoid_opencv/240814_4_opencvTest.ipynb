```python
import cv2
import matplotlib.pyplot as plt

# 이미지 경로
image_path = './img/cat3.jpg'

# 이미지 읽기
img = cv2.imread(image_path)

# 이미지가 제대로 로드되었는지 확인
if img is not None:
    # BGR을 RGB로 변환 (Matplotlib는 RGB 형식을 사용)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    # Matplotlib를 사용하여 이미지 표시
    plt.imshow(img_rgb)
    plt.axis('off') # 축 숨기기
    plt.show()
else:
    print("이미지를 불러올 수 없습니다. 경로를 확인하세요.")
```
