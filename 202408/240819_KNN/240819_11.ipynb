```python
import matplotlib.pyplot as plt

# 연도와 해당 연도의 스마트 팩토리 시장 규모 (가상의 예제 데이터)
years = ['2020', '2021', '2022', '2023', '2024', '2025']
market_size = [1000, 1150, 1322.5, 1520.875, 1749.00625, 1600]  # 단위: 억 달러

# 막대 그래프 그리기
plt.figure(figsize=(10, 6))
plt.bar(years, market_size, color='skyblue')

# 그래프 제목 및 축 라벨 설정
plt.title('Smart Factory Market Growth (2020-2025)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Market Size (Billion USD)', fontsize=14)

# 데이터 값 표시
for i, value in enumerate(market_size):
    plt.text(i, value + 20, f'{value:.1f}', ha='center', fontsize=12)

# 그래프 출력
plt.ylim(0, max(market_size) * 1.2)  # y축의 범위를 조정하여 텍스트가 겹치지 않게 설정
plt.show()
```
