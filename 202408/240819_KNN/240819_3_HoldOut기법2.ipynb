```python
X = df.iloc[:, :60].values
y = df.iloc[:, 60].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 기본값 훈련셋, 테스트셋
# train_test_split ---> HoldOut 
# 0.75 : 0.25 비율이 기본값이다. (*100 퍼센트)
# 섞여서 나온다.
```

```python
train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
X_train , X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.3, shuffle=True)
```

```python
# 학습용, 테스트용
# X_train , X_test, 
# y_train , y_test
```

```python
model = Sequential()
model.add( Dense(24, 
                input_dim=60,
                activation='relu'
))
model.add( Dense(10, activation='relu'))
model.add( Dense(1, activation='sigmoid'))

model.compile( 
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

history = model.fit( X_train, y_train,
                    epochs=200, batch_size=10 )
```

```python
model = Sequential()
model.add( Dense(24, 
                input_dim=60,
                activation='relu'
))
model.add( Dense(10, activation='relu'))
model.add( Dense(1, activation='sigmoid'))

model.compile( 
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

history = model.fit( X_test, y_test,
                    epochs=200, batch_size=10 )
```

```python
# 실제 모델 정확도는?? 테스트용으로 확인 해라. 
# model.evaluate( x=None,  
#     y=None,  
#     batch_size=None,  
#     verbose='auto',  
#     sample_weight=None,  
#     steps=None,  
#     callbacks=None,  
#     max_queue_size=10,  
#     workers=1,  
#     use_multiprocessing=False,  
#     return_dict=False,  
#     **kwargs,  )
model.evaluate( X_test, y_test )
```
