```python
X = data_set.values[:,:8]
y = data_set.values[:,8]
```

```python
model = Sequential()
model(Dense(12, 
            input_dim = 8,
            activation='relu',
            name="Dense_1"            
))
model(Dense(8,
            activation='relu',
            name="Dense_2"            
))
# model(Dense(8,
#             activation='relu',
#             name="Dense_3"            
# ))
# model(Dense(8,
#             activation='relu',
#             name="Dense_4"            
# ))
# model(Dense(8,
#             activation='relu',
#             name="Dense_5"            
# ))
# model(Dense(8,
#             activation='relu',
#             name="Dense_6"            
# ))
# model(Dense(8,
#             activation='relu',
#             name="Dense_7"            
# ))
model.add(Dense(1, 
                activation='sigmoid',
                name="Dense_10"
            ))
model.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=['accuracy']
)
model.fit(X, y, epochs=100, batch_size=32)
```

```python
model.summary()
```

```python
model2 = Sequential()
model2(Dense(12, 
            input_dim = 8,
            activation='relu'         
))
model2(Dense(8,
            activation='relu'         
)) 
model2.add(Dense(1, 
                activation='sigmoid' 
            ))
model2.compile(
    loss="binary_crossentropy",
    optimizer="adam",
    metrics=['accuracy']
)
model2.fit(X, y, epochs=100, batch_size=5)
```
