```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import matplotlib.pyplot as plt
# OpenCV   
image = np.zeros((300, 600, 3), dtype=np.uint8)
image[:] = (255, 255, 255)  #  
# matplotlib   
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False  #     
plt.figure(figsize=(6, 3))
plt.text(0.5, 0.5, ', OpenCV!', fontsize=20, ha='center', va='center')
plt.axis('off')
#   
plt.draw()
text_image = np.frombuffer(plt.gcf().canvas.tostring_rgb(), dtype=np.uint8)
text_image = text_image.reshape(
    plt.gcf().canvas.get_width_height()[::-1] + (3,))
# OpenCV    
text_image = cv2.cvtColor(text_image, cv2.COLOR_RGB2BGR)
image = cv2.addWeighted(image, 0.5, text_image, 0.5, 0)
#   
cv2.imshow("Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
from PIL import ImageFont, ImageDraw, Image

#    ( )
image = np.zeros((500, 500, 3), dtype="uint8")
image.fill(255)  #  

# OpenCV  Pillow  
pil_image = Image.fromarray(image)

#   (    )
fontpath = "./MALGUN.TTF"  # TTF    
font = ImageFont.truetype(fontpath, 40)  #   40 

# ImageDraw  
draw = ImageDraw.Draw(pil_image)

#   
text = " "
draw.text((50, 200), text, font=font, fill=(0, 0, 0, 0))  # (x, y)   

# Pillow   OpenCV  
image = np.array(pil_image)

#  
cv2.imshow("Image with Korean Text", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

file_path = './img/girl.jpg'
img = cv2.imread(file_path)                            #    
img_gray = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE) #    

cv2.namedWindow('origin')                               # origin   
cv2.namedWindow('gray', cv2.WINDOW_NORMAL)              # gray   
cv2.imshow('origin', img)                               # origin   
cv2.imshow('gray', img_gray)                            # gray   

cv2.moveWindow('origin', 0, 0)                          #   
cv2.moveWindow('gray', 100, 100)                        #   

cv2.waitKey(0)                                          #  
cv2.resizeWindow('origin', 200, 200)                    #    ( )
cv2.resizeWindow('gray', 100, 100)                      #    ( ))

cv2.waitKey(0)                                          #  
cv2.destroyWindow("gray")                               # gray  

cv2.waitKey(0)                                          #  
cv2.destroyAllWindows()                                 #   
```
