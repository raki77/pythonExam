```python
# norm : 크기 , 노말라이징 -> 단위벡터 구하기
# norm : 규격, 기준
# 결국은 np.sqrt(np.sum(vec1**2)) 이런 형식이다.
# 단위벡터는 결국 x/norm , y/norm, z/norm이다.

import numpy as np
from scipy import linalg
```

```python

# rank (차원)
x1 = np.array([[2,7],
              [5,1]])
```

```python
x1
```

```python
x1.ndim
np.linalg.matrix_rank(x1)
```

```python
A1 = np.array([[1,2], [3,5]])
B1 = np.array([[1,2], [3,6]]) 
```

```python
np.linalg.matrix_rank(A1)
```

```python
# 선형 종속이라 1이 출력된다.
np.linalg.matrix_rank(B1)
```

```python
A1 = np.array([[1,2], [3,4]])
B1 = np.array([[5,6], [7,8]]) 
```

```python
# 행렬의 곱
A1.dot(B1)
```

```python
#결과가 스칼라로 나온다.
np.dot(A1, B1)
```
