```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
import pandas as pd
```

```python
df_ori = pd.read_csv('./data/sonar3.csv',header=None)
df = df_ori
```

```python
df.head()
```

```python
df.info()
```

```python
df[60].value_counts()
```

```python
X = df.iloc[:, :60].values
y = df.iloc[:, 60].values
```

```python
X
```

```python
y
```

```python
model = Sequential()
model.add( Dense(24, 
                input_dim=60,
                activation='relu'
))
model.add( Dense(10, activation='relu'))
model.add( Dense(1, activation='sigmoid'))

model.compile( 
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

history = model.fit(X, y,
                    epochs=200,
                    batch_size=10)

# accuracy=100 이 나와서, 과적합 되었다고 판단한다.
```
