```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr 
```

```python
data = pd.DataFrame({'x': bream_length, 'y': bream_weight})

corr_coef, p_value = pearsonr(data['x'], data['y'])
print(f"피어슨 상관계수 (scipy): {corr_coef:.3f}, p-value: {p_value:.3f}")

corr_matrix = data.corr(method='pearson')
print("피어슨 상관계수 (pandas):")
print(corr_matrix)
```

```python
data.corr()
```

```python
# 피어슨 상관 계수
pearson_corr = data.corr(method='pearson')

# 스피어만 상관 계수
spearman_corr = data.corr(method='spearman')

# 켄달 상관 계수
kendall_corr = data.corr(method='kendall')

print("Pearson Correlation:\n", pearson_corr)
print("Spearman Correlation:\n", spearman_corr)
print("Kendall Correlation:\n", kendall_corr)
```
