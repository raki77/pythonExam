```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5,5,0.1)
```

```python
y1 = x**2 - 2
```

```python
y2 = 20 * np.cos(x)**2
```

```python
plt.figure(1, figsize=(5,5))
plt.plot(x,y1, color='r')

plt.figure(2, figsize=(10, 2.5))
plt.plot(x,y2)
plt.show()
```

```python
x = np.arange(0,10,0.1)
y1 = 0.3 * (x-5)**2
y2 = -1.5*x + 3
y3 = np.sin(x)**2
y4 = 10 * np.exp(-x)+1
```

#### limit

```python
# limit
x = np.linspace(-4, 4, 100)
y1 = x**3
y2 = 10 * x**2 - 2

plt.figure(figsize=(6,3))
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()

```

```python
x = np.linspace(-4, 4, 100)
y1 = x**3
y2 = 10 * x**2 - 1

plt.figure(figsize=(6,3))
plt.plot(x, y1, 'o-r', x, y2 , '*-b')
plt.show()
```

```python
# limit -1 ~ 1     .   
x = np.linspace(-4, 4, 100)
y1 = x**3
y2 = 10 * x**2 - 1

plt.figure(figsize=(6,3))
plt.plot(x, y1, 'o-r', x, y2 , '*-b')
plt.xlim(-1, 1)
plt.ylim(-3, 3)
plt.xlabel('x')
plt.ylabel('y')
plt.title('My plot')
plt.legend(['y1=$x^3$', 'y2=$10x^2-1$'],loc=1) # 1 : 'upper right'
plt.show()
```
