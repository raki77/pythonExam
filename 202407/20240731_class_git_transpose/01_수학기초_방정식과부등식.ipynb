# 연립 방정식:
###  미지수가 여러개 포함된 방정식을 묶어 놓은 것을 의미한다.
### 아래는 연립방정식이다.

$ 3x + y = 2$      

$ x - 2y = 3$




```python
x = Symbol('x')
y = Symbol('y')

eq1 = 3 * x + y - 2
eq2 = x - 2 * y - 3

solve((eq1, eq2))
```
