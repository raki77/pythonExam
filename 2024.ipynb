```python
# 컬렉션 관리
# 기타 기능들
# GUI (Tkinter)
# DB연동

# enumerate : 시퀀스의 인덱스와 요소를 튜플로 묶어서 순회한다.

# 하나의 튜플에 더 많은 항목이 포함된 경우
race = ['저그', '테란', '프로토스']
print(list(enumerate(race)))

score = [88, 95, 70, 100, 99]
for no, s in enumerate(score, 1):
    print(f"{no}번 학생의 성적 : {s}")

# 하나의 튜플에 더 많은 항목이 포함된 경우
numbers = [1,2,3]
letters = ["A", "B", "C"]
for pair in zip(numbers, letters):
    print(pair)

# 컬렉션 사본
# 시퀀스.copy() -> 시퀀스 복사본
# listcopy1, listcopy2

list1 = [1,2,3]
list2 = list1 #참조
print(list1 == list2)
list2[1] = 100
print(list1)
print(list2)
print(list1==list2)

# listcopy2
list1 = [1,2,3]
list2 = list1.copy() # list2=list1[:] 동일하다.

print(list1 == list2)
list2[1] = 100
print(list1)
print(list2)
print(list1 == list2)

# is 연산자
# == 연산자(상등 연산자)는 변수의 값이 동일한 지를 조사하는 연산자
# is 연산자는 동일한 객체를 가리키는지 조사하는 연산자

s1 = 'bbbb'
s2 = 'ccc'
print(s1 is s2)


# Filter (함수, 리스트나 튜플)
# 일반 함수 버젼
def is_even(x):
    return x % 2 == 0
result1 = list(filter(is_even, range(10)))
print(result1)


# 람다 함수 버젼
# 필터 : 첫번째 인자에는 두번째 인자로 들어온 리스트나 튜플을 하나하나 씩 받아서
# 필터링 한다.
result2 = list( filter((lambda x:x%2==0), range(10)))
print(result2)

# map (함수, 리스트나 튜플) 
# 버젼 첫 번째 인자인 함수는 두 번째 인자로 들어온 리스트나, 튜플에 특별한 
# 가공 처리를 하는 함수

# 일반 함수 버젼
def plus_two(x):
    return x+2
result1 = list(map(plus_two, [1,2,3,4,5]))
print(result1)

#람다 함수 버젼
result2 = list(map((lambda x:x+2), [1,2,3,4,5]))
print(result2)


# 람다 함수
# lambda 인자 : 표현식, 이름없는 함수

def is_even(x):
    return x%2 == 0

is_even = lambda x: x%2 == 0

#정규식 
# re 모듈 사용, 정규 표현식 활용, 문자열에서 패턴 찾고, 다양한 작업 가능
import re
text = "The cat and the hat"
pattern  = r'\b\w+at\b'
result = re.findall(pattern, text)
print(result)



```
