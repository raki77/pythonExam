```python
titanic = titanic.dropna(axis=0)
```

```python
print("결측치 확인:", titanic.isna().sum())
```

```python
titanic.head()
```

```python
titanic['sex'] = titanic.iloc[ titanic['sex'] == "male"] = 0
titanic['sex'] = titanic.iloc[ titanic['sex'] == "female"] = 1 
titanic.head()
```

```python
# titanic['ipt'] if titanic[(titanic['age'] > 50) & (titanic['sex'] == 'female')] : 'weak' else 'strong' 
# 조건에 따라 'ipt' 열에 값 할당, 0노약자, 1젊은이
titanic['ipt'] = np.where(
    (titanic['age'] > 60) & (titanic['sex'] == 1), 0, 1
) 
df_output = titanic['survived'] 
X = titanic.iloc[:,:6]
y = titanic['survived']
```

```python
titanic['ipt']
```

```python
from sklearn.model_selection import train_test_split
import seaborn as sns
import pandas as pd

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2, test_size=0.4)
```

```python
print(f"훈련 데이터 크기: X_train {X_train.shape}, y_train {y_train.shape}")
print(f"테스트 데이터 크기: X_test {X_test.shape}, y_test {y_test.shape}")
```

```python
# KNN 모델 생성 및 훈련
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=9)
model.fit(X_train, y_train)

# 예측 및 결과 출력
pred = model.predict(X_test)
print("예측값:", pred)
print("실제값:", list(y_test))

# 정확도 출력
print("train score:", model.score(X_train, y_train))
print("test score:", model.score(X_test, y_test)) 
```

```python
print(list(titanic['ipt']))
class_names = list(titanic['ipt']) 
for i, j in zip(pred, y_test) :
    print("예측 : {:20f} 실제: {:20f}".format(class_names[i], class_names[j])) 
```
