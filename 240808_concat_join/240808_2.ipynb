```python

import numpy as np
from scipy import linalg
import pandas as pd
```

```python
dic1 = {
    'Name' : ['David Baker',
        'John Smith',
        'Juan Martinez',
        'Jane Connor'
    ],
    'Gender' : [
        'Male',
        'Male',
        'Male',
        'Female'
    ],
    'Age' :[
        31,
        22,
        36,
        30
    ]
}

dic2 = {
    'Name' : [
        'John Smith',
        'Alex Du Bois',
        'Joanne Rowling',
        'Jane Connor'
    ],
    'Position' : [
        'Intern',
        'Team Lead',
        'Manager',
        'Manager'
    ],
    'Wage' : [
        25000,
        75000,
        90000,
        70000      
    ]    
}
```

```python
df1 = pd.DataFrame(dic1)
df1
```

```python
df2 = pd.DataFrame(dic2)
df2
```

```python
# df1.join(pd2)
# help(df1.join)

# left 기준으로 join : left join
# right 기준으로 join  : right join
# inner join
# outer join
```

```python
# inner join
pd.merge(df1, df2)
```
