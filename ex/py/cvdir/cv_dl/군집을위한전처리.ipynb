```python
abs_diff = np.abs(fruits - banana_mean)
```

```python
abs_diff
```

```python
abs_diff[1:2, :, :]
```

```python
abs_mean = np.mean(abs_diff, axis=(1, 2))

# 3차원 배열은 일반적으로 (depth, height, width)의 형태로 구성되며, 각 차원은 아래와 같은 의미를 가집니다:

# axis=0: 첫 번째 축 (depth 또는 채널 등, 3차원 배열의 첫 번째 차원)
# axis=1: 두 번째 축 (height 또는 행, 2차원에서의 Y축에 해당)
# axis=2: 세 번째 축 (width 또는 열, 2차원에서의 X축에 해당)
```

```python
np.argsort(abs_mean)
```

```python
banana_index = np.argsort(abs_mean)[:100]
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(10, 10, figsize=(10, 10))
for i in range(10):
    for j in range(10):
        axs[i, j].imshow(fruits[banana_index[i*10 + j]], cmap='gray_r')
        axs[i, j].axis('off')
plt.show()
```
