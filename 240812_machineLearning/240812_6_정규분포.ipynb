```python
# 평균이면 m 넣어준다. , 표준편차 >>> sigma 2
sigma = 1.0
mean = 0.0
m = 0
#  (1 / ( SQRT(2*PI())*1) )* EXP(1)^-((C36-0)^2 / (2 * 1^2))
y= ( 1.0 / (np.sqrt(2.0 * np.pi * sigma)) ) * np.exp(1.0)**-((x-m)**2.0 / (2.0 * sigma **2.0) )
```
